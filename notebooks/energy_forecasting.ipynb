```python
#Build the robust parser
import pandas as pd
import glob

files = sorted(glob.glob("EIA923*.xlsx"))

master = []

for file in files:

    print("Reading:", file)

    df = pd.read_excel(
        file,
        sheet_name="Page 1 Generation and Fuel Data",
        header=5
    )

    # clean column names
    df.columns = (
        df.columns
          .astype(str)
          .str.replace('\n',' ',regex=False)
          .str.strip()
    )

    # handle old/new fuel classification
    if 'MER Fuel Type Code' in df.columns:
        fuel_col = 'MER Fuel Type Code'
    elif 'AER Fuel Type Code' in df.columns:
        fuel_col = 'AER Fuel Type Code'
    else:
        print("Fuel column not found")
        continue

    keep = [
        fuel_col,
        'Plant State',
        'YEAR'
    ] + months

    df = df[keep]

    long = df.melt(
        id_vars=[
            fuel_col,
            'Plant State',
            'YEAR'
        ],
        value_vars=months,
        var_name='month',
        value_name='generation'
    )

    long = long.rename(
        columns={
            fuel_col:'fuel'
        }
    )

    long['month'] = long['month'].map(month_map)

    master.append(long)

master = pd.concat(master, ignore_index=True)

print(master.shape)
master.head()
```

```python
master['YEAR'] = pd.to_numeric(
    master['YEAR'],
    errors='coerce'
)

master = master.dropna(subset=['YEAR'])

master['YEAR'] = master['YEAR'].astype(int)
```

```python
master['generation'] = pd.to_numeric(
    master['generation'],
    errors='coerce'
)

master = master.dropna(subset=['generation'])

master = master[
    master['generation'] >= 0
]

master = master[
    master['fuel'] != 'HPS'
]
```

```python
#test for stationarity again
from statsmodels.tsa.stattools import adfuller

result = adfuller(
    ng['diff'].dropna()
)

print("ADF:", result[0])
print("p-value:", result[1])
```

```python
#differencing with lag of 12
ng['seasonal_diff'] = ng['generation'].diff(12)

result = adfuller(
    ng['seasonal_diff'].dropna()
)

print("ADF:", result[0])
print("p-value:", result[1])
```

```python
#combined differencing
ng['double_diff'] = (
    ng['generation']
    .diff()
    .diff(12)
)

result = adfuller(
    ng['double_diff'].dropna()
)

print("ADF:", result[0])
print("p-value:", result[1])
```

This is a very common pattern in electricity markets.

Your results indicate that:

The original natural gas generation series is non-stationary.
The series exhibits strong annual seasonality.
Seasonal differencing alone already achieves stationarity.
Applying both regular and seasonal differencing produces an even stronger stationary process.

This suggests that our future forecasting model should be a seasonal ARIMA (SARIMA) model.

```python
#Identify the SARIMA orders
#we need to inspect the autocorrelation structure.
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

series = ng['double_diff'].dropna()

plt.figure(figsize=(12,5))
plot_acf(series, lags=36)
plt.show()

plt.figure(figsize=(12,5))
plot_pacf(series, lags=36)
plt.show()
```

```python
#First task for the Energy Transition Analysis
fuel_mix = (
    master
    .groupby(['YEAR','fuel'])['generation']
    .sum()
    .reset_index()
)

fuel_mix.head()
```

```python
#fuel mix table
fuel_pivot = fuel_mix.pivot(
    index='YEAR',
    columns='fuel',
    values='generation'
).fillna(0)

fuel_pivot.head()
```

```python
#market shares
fuel_share = fuel_pivot.div(
    fuel_pivot.sum(axis=1),
    axis=0
) * 100

fuel_share.head()
```

```python
import matplotlib.pyplot as plt

major = ['COL','NG','NUC','WND','SUN','HYC']

plt.figure(figsize=(14,7))

for fuel in major:
    plt.plot(
        fuel_share.index,
        fuel_share[fuel],
        marker='o',
        label=fuel
    )

plt.title(
    'US Electricity Generation Market Share by Fuel (2015-2024)'
)

plt.xlabel('Year')
plt.ylabel('Market Share (%)')

plt.legend()

plt.grid(True)

plt.show()
```

```python
#growth rates (CAGR)
import pandas as pd

growth = []

for fuel in fuel_pivot.columns:

    start = fuel_pivot.loc[2015, fuel]
    end = fuel_pivot.loc[2024, fuel]

    if start > 0:
        cagr = ((end/start)**(1/9)-1)*100

        growth.append([
            fuel,
            start,
            end,
            cagr
        ])

growth = pd.DataFrame(
    growth,
    columns=[
        'fuel',
        '2015',
        '2024',
        'CAGR'
    ]
)

growth = growth.sort_values(
    'CAGR',
    ascending=False
)

print(growth)
```

```python
#quantify how diversified the electricity portfolio became.
hhi = []

for year in fuel_share.index:
    shares = fuel_share.loc[year] / 100
    hhi.append(
        [
            year,
            (shares**2).sum() * 10000
        ]
    )

import pandas as pd

hhi = pd.DataFrame(
    hhi,
    columns=['YEAR','HHI']
)

print(hhi)
```
